```python
sub_1 = df_times[['world_rank', 'country', 'teaching', 'international', 'research', 'citations', 'income', 'num_students', 'student_staff_ratio', 'international_students', 'female_male_ratio']]
```

```python
# deleting all rows where theres a "-" in world_rank in sub_1
sub_2 = sub_1.loc[sub_1['world_rank'].str.contains("-", case=False)]
```

```python
# save dataframe in sub_clean - only the rows where 'world_rank' DOESN't contain "-"
sub_clean = sub_1[sub_1['world_rank'].str.contains("-", case=False) == False]
```

```python
# check if all 'world_rank' entries that contain "-" were deleted
sub_clean.loc[sub_clean['world_rank'].str.contains("-", case=False)]
```

```python
# see where 'world_rank' = "=39"
sub_clean.loc[sub_clean['world_rank'].str.contains("=", case=False)]
```

```python
# save dataframe in sub_clean - only the rows where 'world_rank' DOESN't contain "="
sub_clean = sub_clean[sub_clean['world_rank'].str.contains("=", case=False) == False]
```

```python
# see where 'world_rank' = "=39"
sub_clean.loc[sub_clean['world_rank'].str.contains("=", case=False)]
```

```python
# add labels, a legend, and change the size of the heatmap
f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(sub_clean.corr(), fignum=f.number) # type of plot
plt.xticks(range(sub_clean.shape[1]), sub_clean.columns, fontsize=14, rotation=45) # x axis labels
plt.yticks(range(sub_clean.shape[1]), sub_clean.columns, fontsize=14) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix Times', fontsize=14) # add title
```

#### Observation
What's interesting is that there are some strong negative realtionships between international and world_rank, country, teaching. Also between teaching and world_rank/country ranking there is - as expected - a positive relationship. And world_rank and country ranking have almost no relationship, which also makes sense as a university can be the best of a country which doesn't mean that it's best in the world too.

```python
# converting 'world_rank' to float so that it has the same datatype as teaching
sub_clean['world_rank'] = sub_clean['world_rank'].astype(float)
print(sub_clean.dtypes)
```

```python
sns.lmplot(x = 'world_rank', y = 'teaching', data = sub_clean)
```

#### Observation
We can see that the higher the ranking is (meaning lower the number) the higher the university is scored in teaching. What is particularly interesting is that there are a few unis that that have a high score in teaching but a rather low ranking in the world.

```python
# converting 'world_rank' to int
sub_clean = sub_clean.astype({'world_rank':'int'})
```

```python
sub_clean.loc[sub_clean['world_rank'] <= 3, 'world_rank_position'] = 'Top 3'
```

```python
sub_clean.loc[(sub_clean['world_rank'] > 3) & (sub_clean['world_rank'] <= 10), 'world_rank_position'] = 'Top 10'
```

```python
sub_clean.loc[(sub_clean['world_rank'] > 10) & (sub_clean['world_rank'] <= 20), 'world_rank_position'] = 'Top 20'
```

```python
sub_clean.loc[(sub_clean['world_rank'] > 20) & (sub_clean['world_rank'] <= 50), 'world_rank_position'] = 'Top 50'
```

```python
sub_clean.loc[(sub_clean['world_rank'] > 50) & (sub_clean['world_rank'] <= 100), 'world_rank_position'] = 'Top 100'
```

```python
sub_clean['world_rank_position'].value_counts(dropna = False)
```

```python
sns.set(style="ticks")
g = sns.catplot(x="world_rank", y="teaching", hue="world_rank_position", data=sub_clean)
```

#### Observation
As expected we can see that the higher the score in teaching is the higher the university scores overall (`world_rank`). The coloring of the`world_rank_position` helps to read the plot faster.
